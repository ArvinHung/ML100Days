```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# 基礎題
```

```python
# Q1 畫出 cos 圖檔，並儲存

x = np.arange(0,180)
y = np.cos(x*np.pi/180)

plt.figure(figsize=(8,6), facecolor="#CDA581")

plt.plot(x,y)

plt.savefig("H1,Q1.jpg", dpi = 300, format = "jpg")
```

```python
# Q2 給定散點圖顏色

n = 1024

x = np.random.normal(0,1,n)
y = np.random.normal(0,1,n)

T = np.arctan2(y,x)

plt.figure(figsize=(8,6), facecolor="#CDA581")
plt.scatter(x,y, s = 50, c = T, alpha = 0.75)
# c = T >> 使得顏色會變化

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.show()
```
