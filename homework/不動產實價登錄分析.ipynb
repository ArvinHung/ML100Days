## 8."建物型態": 把住宅和華廈合成一個類別，並且針對"建物型態"設 dummy

```python
# 先用 .value_counts() 檢查這個欄位裡面有什麼東西
data_TPE_selc["建物型態"].value_counts()

# 把住宅和華廈合成一個類別
  # 用 ｢看看某某欄位裡面有沒有特定字串」去做
  # 用 if x in df["xxx"] : 
  #    return "住宅大樓與華廈"
def trans_to_cate1(df): 
  if "公寓" in df["建物型態"]:
    return "公寓"

  elif "華廈" in df["建物型態"]:
    return "華廈與住宅大樓"

  elif "住宅大樓" in df["建物型態"]:
    return "華廈與住宅大樓"

  elif "透天厝" in df["建物型態"]:
    return "透天厝"

data_TPE_selc["建物型態_整理後"] = data_TPE_selc.apply(trans_to_cate1, axis = 1)

# 檢查新欄位裡的東西
print( data_TPE_selc["建物型態_整理後"].value_counts(), "\n")

# 針對"建物型態"設 dummy
data_TPE_selc = pd.get_dummies(data_TPE_selc, columns = ["建物型態_整理後"])
data_TPE_selc.columns
```

```python
# 看看裡面有幾種東西
data_TPE_selc["移轉層次"].value_counts()
```

```python
def floor_trans_to_dumber(df): 
    if df["移轉層次"] == "一層": 
        return 1 
    elif df["移轉層次"] == "二層":
        return 2
    elif df["移轉層次"] == "三層":
        return 3
    elif df["移轉層次"] == "四層":
        return 4
    elif df["移轉層次"] == "五層":
        return 5
    elif df["移轉層次"] == "六層":
        return 6        
    elif df["移轉層次"] == "七層":
        return 7
    elif df["移轉層次"] == "八層":
        return 8
    elif df["移轉層次"] == "九層":
        return 9
    elif df["移轉層次"] == "十層":
        return 10
    elif df["移轉層次"] == "十一層":
        return 11
    elif df["移轉層次"] == "十二層":
        return 12
    elif df["移轉層次"] == "十三層":
        return 13
    elif df["移轉層次"] == "十四層":
        return 14
    elif df["移轉層次"] == "十五層":
        return 15
    elif df["移轉層次"] == "十六層":
        return 16
    elif df["移轉層次"] == "十七層":
        return 17
    elif df["移轉層次"] == "十八層":
        return 18
    elif df["移轉層次"] == "十九層":
        return 19
    elif df["移轉層次"] == "二十層":
        return 20
    elif df["移轉層次"] == "二十一層":
        return 21
    elif df["移轉層次"] == "二十二層":
        return 22
    elif df["移轉層次"] == "二十三層":
        return 23
    elif df["移轉層次"] == "二十四層":
        return 24
    elif df["移轉層次"] == "二十五層":
        return 25
    elif df["移轉層次"] == "二十六層":
        return 26
    elif df["移轉層次"] == "二十七層":
        return 27
    elif df["移轉層次"] == "二十八層":
        return 28
    elif df["移轉層次"] == "二十九層":
        return 29
    elif df["移轉層次"] == "三十層":
        return 30  
    elif df["移轉層次"] == "全":
        return "透天厝"
    else:
      return "other"

data_TPE_selc["移轉層次_數值"] = data_TPE_selc.apply(floor_trans_to_dumber, axis = 1)
```
