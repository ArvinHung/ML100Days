```python
# 請根據提供的資料，選擇出下列的要求：

import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))
print(df,"\n")

# - 1. filtered by first column > 20?
a = df.iloc[:,0]
condition = (a>20)
print(a,"\n")
print(a[condition],"\n")

# - 2. filtered by first column + second column > 50
b = df.iloc[:,0]
c = df.iloc[:,1]
d = df.iloc[:,[0,1]]
condition = ((b+c)>50)
#print(condition)
print(d,"\n")
print(d[condition],"\n")

# - 3. filtered by first column < 30 "or" second column > 30
condition = ((b<30)|(c>30))
print(d[condition],"\n")

# - 4. filtered by total sum of row > 100
#法一
f = df[0]+df[1]+df[2]+df[3]
condition = (f>100)
print(df[condition],"\n")

#法二
df["sum"] = df[0]+df[1]+df[2]+df[3]
e = df["sum"]
condition = (e>100)
print(df[condition],"\n")
```
