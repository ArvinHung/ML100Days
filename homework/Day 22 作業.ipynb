```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('titanic')
df
```

```python
# Q1 將每個class中男性和女性的平均存活率用長形圖可視化
plt.figure(figsize=(8,6), facecolor="#CDA581")
sns.barplot(x="sex", y="survived", hue="class", data=df)

# 在x軸 以性別分類
# 在y軸 顯示存活率
# 以顏色 將各性別中的各個艙等(class)分類
```
